```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./shapes.png")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray=  cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_rgb)
```

```python
gray = cv2.Canny(image_gray,100,150) #expect gray image
plt.imshow(gray,cmap = 'grey')
```

```python
contours, hierarchy = cv2.findContours(gray,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
#mode - hierarchy relationship between contours
#meathod - contours points are stored

contour_image = image.copy()
cv2.drawContours(contour_image,contours,-1,(0,255,0),5)

plt.figure(figsize= (10,5))
plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(gray,cmap= 'grey')
plt.title("maks")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image,cv2.COLOR_BGR2RGB))
plt.title("Contours image")

plt.show()
```

```python
#Using inrange meathod
import numpy as np
lower_bound = np.array([240], dtype = np.uint8)
upper_bound = np.array([255], dtype = np.uint8)

image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
image_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
mask = cv2.inRange(image_gray,lower_bound,upper_bound)

plt.imshow(mask,cmap='grey')
contours, hierarchy = cv2.findContours(mask,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)

sorted_contours = sorted(contours, key = cv2.contourArea,reverse=True)
contours = sorted_contours[1:6]
#remove the image boundary
```

```python
contour_image = image.copy()
for contour in contours:
    x,y,w,h = cv2.boundingRect(contour)
    cv2.rectangle(contour_image,(x,y),(x+w,y+h),(0,0,255),4)


# cv2.drawContours(contour_image,contours,-1,(0,255,0),5) # -1 for filling at last

plt.figure(figsize= (10,5))
plt.subplot(1,3,1)
plt.imshow(image_gray, cmap = 'grey')
plt.title("Original Image")

plt.subplot(1,3,2)
plt.imshow(mask,cmap= 'grey')
plt.title("maks")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(contour_image,cv2.COLOR_BGR2RGB))
plt.title("Contours image")

plt.show()
```
