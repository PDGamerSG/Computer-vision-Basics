```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create a numpy canvas
height,width = 500,700
blue = (255,127,0)
canvas = np.full((height,width,3),blue,dtype=np.uint8)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
#plt expect rgb photo only
```

Draw a line

```python
cv2.line(canvas,(50,50),(650,450),(0,0,255),5) # first two is y and x
#y is horizontal and x is vertical and after that is color in bgr
#then last is thickness
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

Draw a Rectangle

```python
cv2.rectangle(canvas,(100,100),(300,300),(0,255,255),-1)
#if -1 then it will fill the rectangle
#other +ve is just boundary
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

Draw a Circle

```python
cv2.circle(canvas,(500,150),100,(0,0,255),5)
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

Draw an Ellipse

```python
cv2.ellipse(canvas,(350,350),(100,50),45,0,360,(0,255,0),3)
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```

Draw a Polygon

```python
# Xn, Yn where n>2
points = np.array([[150,400],[300,400],[225,300]],np.int32)
new_reshape_point = points.reshape((-1,1,2))
print(new_reshape_point)

cv2.polylines(canvas,[new_reshape_point],True,(100,255,255),3)
#true and false is for the all the points to be attached thats why
#True means 1,3 point will be attached to make a closed figure
#False means open figure
plt.imshow(cv2.cvtColor(canvas,cv2.COLOR_BGR2RGB))
```
