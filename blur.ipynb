```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./giraffe_kenya.jpg")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
kernel_s = 10
average_blur = cv2.blur(image_grey,(kernel_s,kernel_s))
plt.imshow(average_blur,cmap = 'grey')
```

```python
# after kernel there is sigma x and y in Gausian Blur it is used for
#more smoothness
kernel_s = 31 #it should be odd to work
gaussian_blur = cv2.GaussianBlur(image_grey,(kernel_s,kernel_s),0)
plt.imshow(gaussian_blur,cmap = 'grey')
```

```python
#Median blur only one kernel
median_blur = cv2.medianBlur(image_grey,11)
plt.imshow(median_blur,cmap = 'grey')
```

```python
plt.figure(figsize=(14,12))

plt.subplot(2,2,1)
plt.title("Original greyscale")
plt.imshow(image_grey,cmap = 'grey')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Average Blur")
plt.imshow(average_blur,cmap = 'grey')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Gausian Blur")
plt.imshow(gaussian_blur,cmap = 'grey')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Median Blur")
plt.imshow(median_blur,cmap = 'grey')
plt.axis('off')
```
